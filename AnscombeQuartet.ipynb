### Who is Anscombe?
Francis John "Frank" Anscombe (13th May 1918 - 17th October 2001) was an English statistician.  He studied at Trinity College at Cambridge University where he later returned to lecture.  He also later worked at Princeton and Yale Universities in the US,  where he was elected a Fellow of the American Statistical Association.  He dedicated his entire academic career to the study of statistics. 

As computers became more prominent, he became interested in Statistical computing and made the claim that "a computer should make both calculations and graphs".  To illustrate this claim he compiled four datasets which, although  very produced very similar results when analysed by simple summary satistics but ver different results when graphed.  These dataset together became known as Anscombe's Quartet

### Initial investigatative analysis of the datasets using summary statistics


```python
# to define the dataset parts
x = np.array(aq['x'])
y = np.array(aq['y'])
xI = np.array(x[0:11])
xII = np.array(x[11:22])
xIII = np.array(x[22:33])
xIV = np.array(x[33:44])
yI = np.array(y[0:11])
yII = np.array(y[11:22])
yIII = np.array(y[22:33])
yIV = np.array(y[33:44])

# to calculate and define the summary statistics

# to calculate and define the maximum values for each dataset
xImax = np.max(xI)
xIImax = np.max(xII)
xIIImax = np.max(xIII)
xIVmax = np.max(xIV)
yImax = np.max(yI)
yIImax = np.max(yII)
yIIImax = np.max(yIII)
yIVmax = np.max(yIV)
xImin = np.max(xI)

# to calculate and define minimum values for each dataset
xIImin = np.min(xII)
xIIImin = np.min(xIII)
xIVmin = np.min(xIV)
yImin = np.min(yI)
yIImin = np.min(yII)
yIIImin = np.min(yIII)
yIVmin = np.min(yIV)

# to calculate and define mean values for each dataset
xImean = np.mean(xI)
xIImean = np.mean(xII)
xIIImean = np.mean(xIII)
xIVmean = np.mean(xIV)
yImean = np.mean(yI)
yIImean = np.mean(yII)
yIIImean = np.mean(yIII)
yIVmean = np.mean(yIV)

# to calculate and define standard deviations from mean for each dataset
xIstd = np.std(xI)
xIIstd = np.std(xII)
xIIIstd = np.std(xIII)
xIVstd = np.std(xIV)
yIstd = np.std(yI)
yIIstd = np.std(yII)
yIIIstd = np.std(yIII)
yIVstd = np.std(yIV)

# compute coefficients for simple linear regression for each dataset
tempI = xI - xImean
clI = np.sum(tempI * (yI - yImean)) / np.sum(tempI ** 2)
cOI = yImean - clI * xImean
xI2 = [0,20]
yI2 = [cOI + clI*0, cOI + clI*20]

tempII = xII - xIImean
clII = np.sum(tempII * (yII - yIImean)) / np.sum(tempII ** 2)
cOII = yIImean - clII * xIImean
xII2 = [0,20]
yII2 = [cOII + clII*0, cOII + clII*20]

tempIII = xIII - xIIImean
clIII = np.sum(tempIII * (yIII - yIIImean)) / np.sum(tempIII ** 2)
cOIII = yIIImean - clIII * xIIImean
xIII2 = [0,20]
yIII2 = [cOIII + clIII*0, cOIII + clIII*20]

tempIV = xIV - xIVmean
clIV = np.sum(tempIV * (yIV - yIVmean)) / np.sum(tempIV ** 2)
cOIV = yIVmean - clIV * xIVmean
xIV2 = [0,20]
yIV2 = [cOIV + clIV*0, cOIV + clIV*20]


```

```python
import scipy.stats as stats
fitI = stats.norm.pdf(xI, xImean, xIstd)
plt.plot(xI, fitI, "-o")
plt.hist(xI, normed=True)
plt.show()

fitII = stats.norm.pdf(xII, xIImean, xIIstd)
plt.plot(xII, fitII, "-o")
plt.hist(xII, normed=True)
plt.show()

fitIII = stats.norm.pdf(xIII, xIIImean, xIIIstd)
plt.plot(xIII, fitIII, "-o")
plt.hist(xIII, normed=True)
plt.show()

fitIV = stats.norm.pdf(xIV, xIVmean, xIVstd)
plt.plot(xIV, fitIV, "-o")
plt.hist(xIV, normed=True)
plt.show()

```
