### Initial investigatative analysis of the datasets using summary statistics


```python
# to define the dataset parts
x = np.array(aq['x'])
y = np.array(aq['y'])
xI = np.array(x[0:11])
xII = np.array(x[11:22])
xIII = np.array(x[22:33])
xIV = np.array(x[33:44])
yI = np.array(y[0:11])
yII = np.array(y[11:22])
yIII = np.array(y[22:33])
yIV = np.array(y[33:44])

# to calculate and define the summary statistics
xImax = np.max(xI)
xIImax = np.max(xII)
xIIImax = np.max(xIII)
xIVmax = np.max(xIV)
yImax = np.max(yI)
yIImax = np.max(yII)
yIIImax = np.max(yIII)
yIVmax = np.max(yIV)
xImin = np.max(xI)
xIImin = np.min(xII)
xIIImin = np.min(xIII)
xIVmin = np.min(xIV)
yImin = np.min(yI)
yIImin = np.min(yII)
yIIImin = np.min(yIII)
yIVmin = np.min(yIV)
xImean = np.mean(xI)
xIImean = np.mean(xII)
xIIImean = np.mean(xIII)
xIVmean = np.mean(xIV)
yImean = np.mean(yI)
yIImean = np.mean(yII)
yIIImean = np.mean(yIII)
yIVmean = np.mean(yIV)
xIstd = np.std(xI)
xIIstd = np.std(xII)
xIIIstd = np.std(xIII)
xIVstd = np.std(xIV)
yIstd = np.std(yI)
yIIstd = np.std(yII)
yIIIstd = np.std(yIII)
yIVstd = np.std(yIV)
```

```python
print ("definitions", xImax, xIImax, xIIImax, xIVmax, yImax, yIImax, yIIImax, yIVmax, xImin, xIImin, xIIImin, xIVmin, yImin, yIImin, yIIImin, yIVmin, xImean, xIImean, xIIImean, xIVmean, yImean, yIImean, yIIImean, yIVmean, xIstd, xIIstd, xIIIstd, xIVstd, yIstd, yIIstd, yIIIstd, yIVstd)
```
